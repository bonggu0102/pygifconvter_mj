```python
# 활동 지역 명사 빈도
okt = Okt()
df_work_area_1 =df[['WORK_AREA_CMMN_CODE_SE_1_1','WORK_AREA_CMMN_CODE_SE_1_2','WORK_AREA_CMMN_CODE_SE_2_1','WORK_AREA_CMMN_CODE_SE_2_2']]
df_work_area_1 =df_work_area_1.fillna("")
df_work_area_1['WORK_AREA_CMMN_CODE_SE'] = df_work_area_1['WORK_AREA_CMMN_CODE_SE_1_1']+df_work_area_1['WORK_AREA_CMMN_CODE_SE_1_2']
df_work_area_1['WORK_AREA_CMMN_CODE_SE_2'] = df_work_area_1['WORK_AREA_CMMN_CODE_SE_2_1']+df_work_area_1['WORK_AREA_CMMN_CODE_SE_2_2']
df_work_area_2 = df_work_area_1[['WORK_AREA_CMMN_CODE_SE','WORK_AREA_CMMN_CODE_SE_2']]
df_work_area_2.head(10)
df_work_area_2['WORK_AREA']=df_work_area_2['WORK_AREA_CMMN_CODE_SE']+" " +df_work_area_2['WORK_AREA_CMMN_CODE_SE_2']
df_work_area_2['WORK_AREA']
df_work_area_2['noun'] = df_work_area_2['WORK_AREA'].apply(lambda x :okt.nouns(x))
df_work_area_2['noun']

all_nouns = sum(df_work_area_2['noun'], [])
count = Counter(all_nouns)
print(count)
nouns_work_area = pd.DataFrame(list(count.items()), columns = ['단어', '빈도'])
nouns_work_area = nouns_work_area.sort_values(by='빈도', ascending=False)
nouns_work_area.head(20)

# 활동 지역의 명사들을 추출해서 빈도를 세어보았는데 서울이 6331번으로 제일 많았고 강서구가 311번으로 두번째로 많았다.
# 그 다음으로는 중랑구(270번), 강남구(241번), 노원구(228번)이 뒤를 이었다.
# 따라서 이 데이터들을 통해 사람들이 근무하는 지역 분포가 서울 강서구, 서울 중랑구, 서울 강남구, 서울 노원구 순으로 많은 것을 파악할 수 있었다.
```
